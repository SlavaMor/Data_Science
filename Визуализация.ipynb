# SEABORN

import seaborn as sns

sns.lmplot() - 

Аргументы:  
x - данные по оси Х

y - данные по оси Y

data - данные источник (датафрейм)

hue - разграничивает цветами переменные на графике

fit_reg (если не нужна линия регрессии = False)

Если нужно **поменять размер графика**, то есть надстройка стандартного размера в seaborn

sns.set(rc = {'figure.figsize' : (9, 6)})

```python
ax = sns.lmplot(x= 'math_score', y= 'reading_score', data= df, hue = 'gender', fit_reg= False)
ax.set_xlabels('Math score') # наименование оси Х
ax.set_ylabels('Reading score') # наименование оси Y
```
